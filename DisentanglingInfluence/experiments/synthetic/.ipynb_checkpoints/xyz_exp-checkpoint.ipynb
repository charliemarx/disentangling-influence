```python
import pandas as pd
import numpy as np
from Classifier import classifier
from DisentanglingInfluence.Influence.influence import DR_influence
from DisentanglingInfluence.Disentangling.disentangle import disentangle
from DisentanglingInfluence.utils import DataGenerator
```

```python
output_dir = "../../outputs/synthetic_test/"
```

```python
data = pd.read_csv("../../data/synthetic/DI_synth_xyz.csv")
```

```python
mask = np.random.rand(len(data)) < 0.67
train = data[mask]
test = data[~mask]
```

```python
import shap
from keras.models import Model
from keras.layers import Input, Dense
from keras.optimizers import SGD
```

```python
data.columns.values
```

```python
input_features = ['X','Y','noise']
output_features = ['Z']
```

```python
input_layer = Input(shape=(len(input_features),))
output_layer = Dense(1)(input_layer)
model = Model(inputs=input_layer, outputs=output_layer)
opt = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(opt, loss='mse', metrics=['mae','mse'], loss_weights=None, sample_weight_mode=None, weighted_metrics=None, target_tensors=None)
model.fit(train[input_features],train[output_features], epochs=4, verbose=1)
```

```python
test.loc[0:100][input_features]
```

```python
explainer = shap.KernelExplainer(model.predict, train[input_features], link="logit")
shap_values = explainer.shap_values(np.array(test.loc[0:100][input_features]), nsamples=10)

```
